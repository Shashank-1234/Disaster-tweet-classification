```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Load the TinyBERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("huawei-noah/TinyBERT_General_4L_312D")
model = AutoModelForSequenceClassification.from_pretrained("huawei-noah/TinyBERT_General_4L_312D", num_labels=2)  # 2 labels for disaster and non-disaster
```

```python
from datasets import load_dataset

# Replace 'path_to_your_dataset.csv' with the actual file path
dataset = load_dataset('csv', data_files={'train': 'train.csv', 'test': 'test.csv'})
```

```python
def tokenize_function(examples):
    return tokenizer(examples[' tweet'], padding="max_length", max_length=25, truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
tokenized_datasets
```

```python
model.resize_token_embeddings(len(tokenizer))
```

```python
!export CUDA_LAUNCH_BLOCKING=1
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-4,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=10,
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['test'],
)

# Train the model
trainer.train()
```

```python
# Save the model and tokenizer
model.save_pretrained("./tinybert_disaster_model")
tokenizer.save_pretrained("./tinybert_disaster_model")
```

```python
import pandas as pd
```

```python
quake_1 = pd.read_csv('/content/timesnow_flood.csv')
```

```python
quake_1.head()
```

```python
d = []
for i in range(len(quake_1)):
  d.append(quake_1['Text'][i])
```

```python
d
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load the saved TinyBERT model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("/content/drive/MyDrive/tinybert_disaster_model")
model = AutoModelForSequenceClassification.from_pretrained("/content/drive/MyDrive/tinybert_disaster_model")
```

```python
pred = []
for i in range(len(d)):

  tweet = d[i]

  # Tokenize the tweet
  inputs = tokenizer(tweet, return_tensors="pt", truncation=True, padding="max_length", max_length=128)

  # Perform inference
  outputs = model(**inputs)
  predictions = torch.argmax(outputs.logits, dim=-1)

  # Interpret the prediction
  label = predictions.item()
  pred.append("Disaster-related" if label == 1 else "Not disaster-related")
  #print("Disaster-related" if label == 1 else "Not disaster-related")
```
